```python
#Task:
#Create a line plot showing the temperature variation over a week. The
#x-axis should represent the days of the week, and the y-axis should
#represent the temperature in degrees Celsius. Add appropriate labels
#and a title to the plot

########################
########################
import matplotlib.pyplot as plt
days = [" Saturday" , "Sunday" , "Monday" , "Tuesday" , "Wednesday" , "Thursday" ," Friday"]
temperatures = [20, 22, 19, 10, 18, 20, 25]  
plt.figure(figsize=(10, 5))
plt.plot(days, temperatures, marker='o', linestyle='-', color='green', markersize=8, linewidth=2)
plt.xlabel("Days of the Week")
plt.ylabel("Temperature (°C)")
plt.title("Temperature Variation Over a Week")
plt.xticks(rotation=45)
plt.show()
```
